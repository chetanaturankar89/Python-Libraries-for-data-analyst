# Seaborn

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive, informative, and statistical graphics easily.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# built-in dataset 
df = sns.load_dataset('tips')
df.head()
```

```python
new_df = pd.read_csv('seaborn_practice_dataset.csv')
new_df.head()
```

## Distribution Plots

#### Histogram
-  sns.histplot(data, bins=30):
-  Visualize the distribution of data with histogram.

```python
sns.histplot(new_df['Age'])
plt.show()
```

```python
sns.histplot(df['total_bill'])
plt.show()
```

#### KDEPlot
-  sns.kdeplot(data, fill=True):
-  Plot kernel density estimate to analyze distribution.

```python
sns.kdeplot(new_df['Salary'], fill = True)
plt.show()
```

```python
sns.kdeplot(df['tip'], fill = True)
plt.show()
```

#### Displot
-  sns.distplot(data): Combine histogram and KDE .

```python
sns.displot(new_df['Salary'], kde = True)
plt.show()
```

```python
sns.displot(df['total_bill'], kde = True)
plt.show()
```

#### Rugplot
-  sns.rugplot(data):
-  marginal tick marks to display data points along the axis.

```python
sns.rugplot(new_df['Age'])
plt.show()
```

#### ecdfplot
-  sns.ecdfplot(data):
-  Visualize cumulative distribution of data.

```python
sns.ecdfplot(new_df)
plt.show()
```

```python
sns.ecdfplot(df)
plt.show()
```

#### violinplot
-  sns.violinplot(x=data):
-  Combine KDE and boxplot features in one visualization.

```python
sns.violinplot(x = new_df['Department'], fill = True)
plt.show()
```

```python
sns.violinplot(x = df['size'], fill = True)
plt.show()
```

#### BoxPlot
-  sns.boxplot(data):
-  Summarize data distribution and detect outliers.

```python
sns.boxplot(new_df[['Age','Salary']])
plt.show()
```

```python
sns.boxplot(df)
plt.show()
```

#### jointplot
-  sns.jointplot(x, y, kind='hex'):
-  Show bivariate distribution using hexagonal bins.

```python
sns.jointplot(x = 'Age', y = 'Salary',data = new_df,  kind = 'hex')
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="hex")
plt.show()
```

## Matrix & Heatmaps

#### heatmap
-  sns.heatmap(data, annot=True)
-  Display correlation or pivot tables with annotations.

```python
# works on numerical values only 
values = new_df[['Age','Experience','Salary','Rating']]
corr = values.corr()
sns.heatmap(corr, annot= True, cmap = 'PuBuGn_r')
plt.show()
```

```python
# works on numerical values only 
values = new_df[['Age','Experience','Salary','Rating']]
corr = values.corr()
# Use specific colormaps for emphasis.-- cmap
# Add spacing lines for better readability.-- linewidths
# Normalize heatmaps for better clarity.


sns.heatmap(corr, annot= True, cmap = 'PuBuGn_r', linewidths=0.5, vmin =  0 , vmax = 1 )
plt.show()
```

#### clustermap
-  sns.clustermap(data)
-  Visualize hierarchical clustering of data matrices.

```python
clustermap = new_df[['Age','Experience','Salary','Rating']]
sns.clustermap(clustermap, cmap = 'twilight_shifted' , standard_scale=1)
plt.show()
```

#### sns.heatmap() is the correct function to visualize the correlation matrix.




```python
# works on numerical values only 


# Compute correlation matrix (only numerical columns)
corr_matrix = new_df.corr(numeric_only=True)

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='Set3_r')
plt.title("Correlation Heatmap")
plt.show()
```

#### pairplot
-  sns.pairplot(data)
-  Show pairwise relationships in a dataset.
-  sns.pairplot() only plots numerical columns on the axes (x and y).
-  We can use only one column at a time in the hue parameter of Seaborn plots, including sns.pairplot().



```python
# result  10 only
plt.figure(figsize=[10,6])
sns.pairplot(new_df[['Rating','Age','Salary','Experience']].head(10))
plt.show()
```

```python
# Differentiate data with colors in pair plots
# The hue parameter should reference a column name that exists in  DataFrame
# hue = 'col name' will differentiate data by color 
plt.figure(figsize=[10,6])
sns.pairplot(new_df[['Rating','Age','Salary','Experience','Gender']].head(10), hue = 'Gender')
plt.show()
```

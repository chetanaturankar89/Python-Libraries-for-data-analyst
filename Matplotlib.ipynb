# Matplotlib

Matplotlib is a data visualization library in Python. It lets you create static, animated, and interactive plots for analyzing and presenting data.

```python
# import 
import matplotlib.pyplot as plt
```

## 1. Basic Charts 

### Line Chart 

Draw a line in a diagram from position (0,0) to position (50,250)

```python
import numpy as np

x = np.array([0,6])
y = np.array([50,250])

plt.plot(x,y)
plt.show()
```

Draw two points in the diagram, one at position (1, 3) and one in position (8, 10) with some Enhancement Functions

```python
x = np.array([1,3])
y = np.array([8,10])

# Matplotlib Enhancement Functions
plt.figure(figsize=(8, 6), dpi=80)

plt.title('Point', color = 'red')
plt.xlabel('x-axis',rotation=45,color = 'blue')
plt.ylabel('y-axis',rotation=45,color = 'blue')
plt.legend('Line A')
plt.grid(True) # Add default grid
plt.grid(color = 'pink',linestyle = '--', linewidth = '0.5')  # Customized
plt.xticks(rotation=45)  # Rotate x-axis ticks
plt.yticks(rotation=45)  # Rotate x-axis ticks


plt.plot(x, y, color='green', linestyle='--', linewidth=2, marker='o')

plt.show()
```

```python
# multiple plots
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y)
plt.show()
```

```python
m = np.array([5,6,3,7,9,5])
plt.plot(m,marker = '*')
plt.show()
```

```python
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,marker = 's',color='m')
plt.show()
```

```python
m = np.array([5,6,3,7,9,5])
plt.plot(m,'s:g',linestyle='-.')
# ('s-square,g-green')
plt.show()
```

Line Syntax	Description
-  '-'	Solid line
-  ':'	Dotted line	
-  '--'	Dashed line	
-  '-.'	Dashed/dotted line

```python
# marker size (ms) ,color (mec), markerfacecolor (mfc)
plt.figure(figsize=[12,6])
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
plt.plot(x,y,marker = 'o',color='m',ms = 30, mec= 'b', mfc= 'r')
plt.show()
```

LINESTYEL

```python
# dotted or :
l = np.array([3,5,7,9,10,35])
plt.plot(l,linestyle = ':')
plt.show()
```

```python
# dashed or (--)
l = np.array([3,5,7,9,10,35])
plt.plot(l,ls = '--')
plt.show()
```

```python
# dashdot
l = np.array([3,5,7,9,10,35])
plt.plot(l,linestyle = '-.', color = 'r')
plt.show()
```

```python
# linewidth , color
l = np.array([3,5,7,9,10,35])
plt.plot(l,linestyle = '-', color = 'r', linewidth='08.63')
plt.show()
```

```python
# multiple lines, grid
a = np.array([3,7,8,2,8])
b = np.array([3,4,5,6,7,9])
plt.title('Multiple Lines', loc = 'right')
plt.grid(axis = 'x', color = 'b', linestyle = ':')
plt.grid(axis = 'y', color = 'g', linestyle = '--')
plt.plot(a,color = 'r', marker = 's',ms=20,mec = 'b')
plt.plot(b,color = 'm',marker='o',ms = 20, mec = 'g')
plt.show()
```

Subplots -- With the subplot() function you can draw multiple plots in one figure:

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
# Multiple subplots 
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title('1')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)
plt.title('2')

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)
plt.title('3')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)
plt.title('4')

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)
plt.title('5')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)
plt.title('6')

plt.suptitle("MY Graphs")
plt.show()
```

## 2. SCATTER PLOT

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
# Single scatter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)


plt.show()
```

```python
# Compare scatter plot 
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)

plt.show()
```

```python
# You can include the colormap in the drawing by including the plt.colorbar() statement:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis', )

plt.colorbar()

plt.show()
```

```python
# alpha
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x,y,s = sizes, alpha = 0.5)
plt.show()
```

Combine Color Size and Alpha

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
colors = np.random.randint(13, size=(13))
plt.scatter(x,y,s=sizes,c = colors, alpha = 0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

## 3. Bar Charts

The bar() function takes arguments that describes the layout of the bars.

```python
# plotting bar chart 
a = np.array([10,20,30,40])
b = np.array(['A','B','C','D'])

plt.bar(b,a,color='skyblue')
plt.show()
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
plt.show()
```

```python
# horizontal bar (barh)
a = np.array([10,20,30,40])
b = np.array(['A','B','C','D'])

plt.barh(b,a,color='skyblue')
plt.show()
```

```python
# bar color 
a = np.array([10,20,30,40])
b = np.array(['A','B','C','D'])

plt.bar(b,a,color='#4CAF50')
plt.show()
```

```python
# Bar width (width)
# The default width value is 0.8
a = np.array([10,20,30,40])
b = np.array(['A','B','C','D'])

plt.bar(b,a,color='GreenYellow',width = 0.2)
plt.show()
```

```python
# Bar height (height)
a = np.array(['A','B','C','D'])
b = np.array([10,20,30,40])

# The default height value is 0.8
plt.barh(a,b,height = 0.5)
plt.show()
```

## 4.Histogram

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

```python
# plotting histogram 
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
# plotting with bins 
x = np.random.randn(1000)
plt.hist(x,bins = 10, color = 'pink', edgecolor = 'black')
plt.show()
```

```python
# plotting with beauty
x = np.random.randn(1000)
plt.hist(x,bins = 10, color = 'pink', edgecolor = 'black', linestyle = '-.')
plt.show()
```

## 5. Pie Chart

```python
# With Pyplot, you can use the pie() function to draw pie charts:
a = np.array([5,10,15,20,25,30])
plt.pie(a)
plt.show()
```

```python
# Add labels 
price = np.array([34,123,87,87])
fruits = np.array(['Mango','Apple','Banana','Kiwi'])
plt.pie(price,labels = fruits)
plt.show()
```

```python
# Add startangle 
price = np.array([34,123,87,87])
fruits = np.array(['Mango','Apple','Banana','Kiwi'])
plt.pie(price,labels = fruits, startangle  = 90)
plt.show()
```

```python
# explode -- sep some part 
price = np.array([34,123,87,87])
fruits = np.array(['Mango','Apple','Banana','Kiwi'])
explodes = [0.2,0.1,0.1,0]
plt.pie(price,labels = fruits, explode = explodes, startangle  = 90)
plt.show()
```

```python
# explode -- sep some part with shadow (shadow = True)
price = np.array([34,123,87,87])
fruits = np.array(['Mango','Apple','Banana','Kiwi'])
explodes = [0.2,0.1,0.1,0]
plt.pie(price,labels = fruits, explode = explodes,shadow = True, startangle  = 90)
plt.show()
```

```python
# Add legends 
price = np.array([34,123,87,87])
fruits = np.array(['Mango','Apple','Banana','Kiwi'])
explodes = [0.2,0.1,0.1,0]
plt.pie(price,labels = fruits, explode = explodes,shadow = True, startangle  = 90, autopct='%1.1f%%' )
# autopct<- This shows percentages with 1 decimal)
plt.legend(title = 'Fruits')
plt.show()
```
